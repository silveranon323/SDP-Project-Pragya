```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Amazon-Products.csv")
df.head()
```

```python
#datacleaning starts from here
df["ratings"]=pd.to_numeric(df["ratings"],errors="coerce")
df=df.dropna(subset=["ratings"])
df["ratings"]=df["ratings"].round(1)
print(df["ratings"])
```

```python
rating_counts=df["ratings"].value_counts().sort_index()
rating_counts
```

```python
plt.figure(figsize=(12, 6))
plt.bar(rating_counts.index,rating_counts.values,color="skyblue")
plt.xlabel("Ratings")
plt.ylabel("Number of Products")
plt.title("How Products Are Rated")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# checking for number of missing values in the dataset
df.isnull().sum()
```

```python
#we saw that discount price and actual price still needs to be cleared lets fix
df['discount_price'] = df['discount_price'].replace({'₹': '', ',': ''}, regex=True).astype(float)

# Fill missing 'discount_price' with the median
df['discount_price'] = df['discount_price'].fillna(df['discount_price'].median())

# Repeat the same for 'actual_price' column
df['actual_price'] = df['actual_price'].replace({'₹': '', ',': ''}, regex=True).astype(float)
df['actual_price'] = df['actual_price'].fillna(df['actual_price'].median())
df.isnull().sum()
```

```python
#since the dataset is very big lets first test on 20000 data
df=df.sample(n=20000,random_state=42,).reset_index(drop=True)
df.shape
df.to_csv("finaldataset.csv")
df
```

```python

# Step 1: Sort products by 'no_of_ratings' first (more ratings means more popular),
# then by 'ratings' (higher ratings mean better quality).
popular_products = df.sort_values(by=['no_of_ratings', 'ratings'], ascending=False)

# Step 2: Get the top 10 most popular products
top_10_products = popular_products[['name', 'ratings', 'no_of_ratings']].head(10)

# Step 3: Display the top 10 most popular products
print(top_10_products)
```

```python
top_10_json = top_10_products.to_dict(orient='records')
top_10_json
```

```python
from flask import Flask, jsonify


app = Flask(__name__)
@app.route('/homepage', methods=['GET'])
def homepage():
    return jsonify(top_10_json)

if __name__ == '__main__':
    app.run(debug=True,port=4429)
```
